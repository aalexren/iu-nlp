- Each smiley is a separate token

````python
import sentencepiece as spm

def sentencepiece_wrapper(text: str) -> List[str]:
    spm.SentencePieceTrainer.train('--input=botchan.txt --model_prefix=m --vocab_size=2000')

    # makes segmenter instance and loads the model file (m.model)
    sp = spm.SentencePieceProcessor()
    sp.load('m.model')

    return sp.encode_as_pieces(text)
````

````python
import re

def re_tokenizer(text: str) -> List[str]:
    '''
    1. Each smiley is a separate token
    2. Each hashtag is an individual token. Each user reference is an individual token
    3. If a word has spaces between them then it is converted to a single token
    4. If a sentence ends with a word that legitimately has a full stop (abbreviations, for example), add a final full stop
    5. All punctuations are individual tokens. This includes double-quotes and single quotes also
    6. A URL is a single token
    '''
    re_full_stop = re.compile(r"[.?!]$")
    if not re.search(re_full_stop, text):
        text += '.'

    smiles = r"((?:[:;]|\b[Xx]){1}[ -]?[)PpDd]+(?=\W))"
    hashtags = r"((?<=\s)[#]{1}\w+[^\W#]+)"
    nicknames = r"([@][A-z]{1}[\w_]+[^\W]+)"
    # full_stops = r"" # replace with full stop
    acronyms = r"((?:[a-zA-Z]\.){2,})"
    urls = r'''(https?:\/\/(www[.])?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}[-a-zA-Z0-9()@:%_\+.~#?&\/\/=]*)'''
    rest = r"(\d+[x]\d+)|(\b\d+[$%]?)|(\d?\w+[-—]?\w*\b)"
    punctuation = r'''([.,!'":;?()\[\]><=\-@\+$&*»«])'''
    conds = '|'.join([urls, smiles, hashtags, nicknames, acronyms, rest, punctuation])
    
    re_conds = re.compile(conds, flags= re.UNICODE | re.M)
    re_matches = re.findall(re_conds, text)
    res = [tuple(filter(None, x))[0] for x in re_matches]
    return res
````

````python
import nltk

def nltk_tweet_tokenizer(text: str) -> List[str]:
    tokenizer = nltk.TweetTokenizer()
    return tokenizer.tokenize(text)
````

## Stemming and Lemmatization

Your task is to write two functions: stem and lemmatize. Input is a text, so you need to tokenize it first.

````python
from nltk.stem.snowball import SnowballStemmer

def stem(text: str) -> List[str]:
    tokens = re_tokenizer(text)
    stemmer = SnowballStemmer(language='english')

    res = []
    for token in tokens:
        res.append(stemmer.stem(token))
    
    return res
````

````python
import spacy

nlp = spacy.load('en_core_web_sm') 

def lemmatize(text: str) -> List[str]:
    tokens = re_tokenizer(text)

    res = []
    for token in tokens:
        lemma = nlp(token)
        for lem in lemma:
            res.append(lem.text)
    return res
````

````python
for file in files:
    with open(file, 'r') as f:
        print(f'---------Open {file}---------')
        cnt = 1
        while s := f.readline():
            print(f'[{cnt}] Stemming: {stem(s)}')
            print(f'[{cnt}] Lemmtization: {lemmatize(s)}')
            cnt += 1
````
